**P4b) (2pts)** Basado en la tabla anterior, construya dos nuevas columnas llamadas `density_food_type` y `ratio_review` que contengan la siguiente información:
 
 * `density_food_type`: Representa el cuociente entre le total de restaurants por tipo de comida y ciudad, respecto del total de restaurantes de la ciudad. (`n_rest/total_rest`)
 * `ratio_review`: Representa el cociente entre a valoración del restaurant por tipo de comida y ciudad, respecto de la valoración promedio de los resturants de la misma ciudad. (`review_prom/review_prom_city`)
 
**P4c) (3pts)** Mediante un gráfico de dispersión, muestre la relación entre `density_food_type` y `ratio_review`. Investigue sobre el parámetro `alpha` dentro de la capa geométrica para una mejor visualizaciónd de los puntos.
`````````
##4a.
resumen <- general %>%  #creamos tabla resumen
  inner_join(location, by = c("id_restaurant" = "id_rest"))%>%  #unimos los dos conjuntos
  group_by(city,food_type ) %>%  #agrupamos por ciudad y tipo de comida para obtener sus resumenes
  summarise(n_rest= n(), review_prom = mean(review)) %>% # conteo de restaurantes y promedio por ciudad y tipo de comida
  left_join( # hacemos una union con la otra parte del conjunto, q es un resumen por ciudad (no ciudad y tipo de comida)
    general %>%
    inner_join(location, by = c("id_restaurant" = "id_rest"))%>%
    group_by(city ) %>% # solo ciudad
    summarise(total_rest= n(), review_prom_city = mean(review))
    , key = "city"
  ) 


# ver el resultado
resumen %>%  # filtremos resumen para ver una tabla similar a la imagen
  filter(city %in% c("san francisco", "san jose" , "oakland") & food_type %in% c("chinese", "italian"))


`````````

`````````
## 4b
resumen_4b <- resumen %>%
                mutate( density_food_type = n_rest/total_rest , 
                        ratio_review = review_prom/review_prom_city)

head(resumen_4b)
`````````

`````````
## 4c
## Se peude analizar como la mayoría de los locales tienen calificaciones cercanas al promedio de su locacion, solo hay pocos casos que se escapando e la muestra 
options(repr.plot.width=9, repr.plot.height = 7)


ggplot(resumen_4b , aes(x = ratio_review , y =  density_food_type ))+
  geom_point(alpha = 0.3, # para visualizar superposición de los puntos
             stroke = 0,
             size = 3,
             position = position_jitter(width = 0.01, height = 0.01))+ # agrega un poco de ruido para evitar superposición 
  geom_smooth()
`````````
